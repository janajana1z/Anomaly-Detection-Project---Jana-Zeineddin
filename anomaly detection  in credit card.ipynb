```python
#save the data
labels = data['Class']
new_data  = data.drop('Class', axis = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

#Scale 'Time' and 'Amount'
new_data[['Time', 'Amount']] = scaler.fit_transform(new_data[['Time', 'Amount']])
```

```python
from sklearn.decomposition import PCA

#reduce to 2 components
pca = PCA(n_components=2)
reduced = pca.fit_transform(new_data)
print("Explained variance:", pca.explained_variance_ratio_)
```

```python
# Visualize
plt.figure(figsize=(10, 7))
plt.scatter(reduced[:, 0], reduced[:, 1], c=labels, cmap='coolwarm', alpha=0.5)
plt.title('PCA - 2D Projection of Transactions')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar(label='Class')
plt.show()
```

```python
X = new_data.values
y_true = labels.values      # for evaluation only
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix

# Initialize and fit
iso_forest = IsolationForest(contamination=0.0017, random_state=42)
y_pred_if = iso_forest.fit_predict(X)

# Convert from {-1, 1} to {1, 0} => fraud = 1
y_pred_if = [1 if x == -1 else 0 for x in y_pred_if]

# Evaluation
print("Isolation Forest Results:")
print(confusion_matrix(y_true, y_pred_if))
print(classification_report(y_true, y_pred_if, digits=4))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

def plot_conf_matrix(y_true, y_pred, title):
    disp = ConfusionMatrixDisplay.from_predictions(y_true, y_pred, display_labels=["Non-Fraud", "Fraud"], cmap="Blues")
    plt.title(title)
    plt.show()

plot_conf_matrix(y_true, y_pred_if, "Isolation Forest")

```

```python
# The Isolation Forest model have:
# Precision: High
# Recall: High
# F1 Score: High
# So it's fast and have great baseline
```

```python
# Combine predictions into original DataFrame
df_results = data.copy()
df_results['Prediction'] = y_pred_if  # 1 = predicted fraud, 0 = predicted normal

# Flagged anomalies
anomalies = df_results[df_results['Prediction'] == 1].copy()
```

```python
# let's explore the distribution of key fields like Amount, Time, and PCA features for anomalies.
#Amount of Flagged Transactions
plt.figure(figsize=(10,5))
sns.histplot(anomalies['Amount'], bins=50, kde=True)
plt.title('Transaction Amounts (Flagged Anomalies)')
plt.xlabel('Amount')
plt.ylabel('Count')
plt.show()
```

```python
#Time Patterns
#let's convert Time (seconds since first transaction) to hour of day to find when fraud tends to happen.

# Convert time to hour
df_results['Hour'] = (df_results['Time'] // 3600) % 24
anomalies['Hour'] = (anomalies['Time'] // 3600) % 24

plt.figure(figsize=(10,5))
sns.countplot(x='Hour', data=anomalies)
plt.title('Hour of Day - Flagged Anomalies')
plt.xlabel('Hour (0–23)')
plt.ylabel('Number of Anomalies')
plt.show()
```

```python
#PCA Feature Patterns
for feature in ['V1', 'V2', 'V3']:
    sns.kdeplot(df_results[feature][df_results['Prediction'] == 0], label='Normal', fill=True)
    sns.kdeplot(df_results[feature][df_results['Prediction'] == 1], label='Anomaly', fill=True)
    plt.title(f'Distribution of {feature} - Normal vs Anomalies')
    plt.legend()
    plt.show()
```
